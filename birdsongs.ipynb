```python
import pandas as pd
import matplotlib.pyplot as plt
import librosa, librosa.display
import numpy as np
import math, random
import torch
import copy
from os import path, listdir
from pydub import AudioSegment
from pydub.utils import make_chunks
```

## Read in metadata

```python
metadata_file = pd.read_csv('archive\metadata.csv')
pd.set_option('display.max_columns', None)
```

```python
path = 'archive\mp3'

MP3_list = []
for x in listdir(path):
    MP3_list.append(x)
```

```python
chunk_length = 30000
chunk_list = []
id_list = []
for file in MP3_list:
    filename = path + '\\' + file
    print(filename)
    myaudio = AudioSegment.from_file(filename, 'mp3')
    chunks = make_chunks(myaudio, chunk_length)
    
    for i, chunk in enumerate(chunks):
        chunk_name = 'archive\chunks' + filename[11:-4] + '_' + str(i) + '.mp3'
        if(chunk.duration_seconds == 30.0):
            chunk.export(chunk_name, format = "mp3")
            chunk_list.append(chunk_name)
            id_list.append(metadata_file[metadata_file['Path'] == 'mp3//' + file]['Recording_ID'].iloc[0])
```

```python
bird_names = []
genus_names = []
epi_names = []
for rec_id in id_list:
    bird_names.append(metadata_file[metadata_file['Recording_ID'] == rec_id]['English_name'].iloc[0])
    genus_names.append(metadata_file[metadata_file['Recording_ID'] == rec_id]['Genus'].iloc[0])
    epi_names.append(metadata_file[metadata_file['Recording_ID'] == rec_id]['Specific_epithet'].iloc[0])

chunk_d = {'chunk_name': chunk_list, 'Recording_ID': id_list, 'English_name': bird_names, 'Genus': genus_names, 'Specific_epithet': epi_names}
chunk_dataframe = pd.DataFrame(data=chunk_d)
chunk_dataframe.to_csv('chunk_metadata.csv')
```
