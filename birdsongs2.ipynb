```python
def chunk_to_png(path):    
    hl = 512
    height = 224
    width = 224
    sample, sample_rate = librosa.load(path, sr=None)
    window = sample[0:width*hl]
    sgram = librosa.stft(sample)
    sgram_mag, _ = librosa.magphase(sgram)
    mel_scale_sgram = librosa.feature.melspectrogram(S=sgram_mag, y = window, sr=sample_rate, n_mels = height, hop_length = hl)
    mel_sgram = librosa.amplitude_to_db(mel_scale_sgram, ref=np.min)
    #librosa.display.specshow(mel_sgram1, sr=sample1_rate, fmin=1024, fmax=8192)
    im = Image.fromarray(mel_sgram)
    im = im.convert('RGB')
    im = im.resize((224, 224))
    im.save('D:\\birdsongs_images' + path[14:-4] + '.png')
    im.close()
```

```python
chunk_list = []
for x in listdir('archive\chunks'):
    chunk_list.append(x)
```

```python
for file in chunk_list:
    chunk_to_png('archive\chunks\\' + file)
```
